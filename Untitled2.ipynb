```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
import gradio as gr

# Sample dataset
data = {
    'message': [
        "Win a free iPhone!",
        "Meeting at 3 PM",
        "Claim your $1000 prize now!",
        "Can we talk later?",
        "Congratulations! You've won a lottery!"
    ],
    'label': ['spam', 'not spam', 'spam', 'not spam', 'spam']
}
df = pd.DataFrame(data)

# Preprocessing
X = df['message']
y = df['label']
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(X)

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model = MultinomialNB()
model.fit(X_train, y_train)

# Function to classify messages
def classify_message(message):
    transformed_message = vectorizer.transform([message])
    prediction = model.predict(transformed_message)
    return f"The message is classified as: {prediction[0]}"

# Create Gradio interface
interface = gr.Interface(
    fn=classify_message,
    inputs=gr.Textbox(label="Enter a Message"),
    outputs=gr.Text(label="Prediction"),
    title="Spam Detection",
    description="Enter a message to check if it is spam or not spam."
)

# Launch the Gradio app
interface.launch()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import gradio as gr

# Function to perform K-means clustering
def perform_clustering(file, num_clusters):
    try:
        # Load customer data from uploaded file
        data = pd.read_csv(file.name)

        # Check for necessary columns
        if 'Annual_Spend' not in data.columns or 'Frequency' not in data.columns:
            return "Error: CSV file must contain 'Annual_Spend' and 'Frequency' columns."

        # Select relevant features
        features = data[['Annual_Spend', 'Frequency']]

        # Normalize the features
        scaler = StandardScaler()
        scaled_features = scaler.fit_transform(features)

        # Apply K-means clustering
        kmeans = KMeans(n_clusters=num_clusters, random_state=42)
        data['Cluster'] = kmeans.fit_predict(scaled_features)

        # Save the segmented data to a new file
        output_file = "segmented_customers.csv"
        data.to_csv(output_file, index=False)
        return output_file
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Gradio Interface
file_input = gr.File(label="Upload CSV File (Include 'Annual_Spend' and 'Frequency')")
num_clusters_input = gr.Slider(2, 10, step=1, value=3, label="Number of Clusters")
output_file = gr.File(label="Download Segmented CSV File")

interface = gr.Interface(
    fn=perform_clustering,
    inputs=[file_input, num_clusters_input],
    outputs=output_file,
    title="Customer Segmentation using K-means",
    description="Upload a CSV file with 'Annual_Spend' and 'Frequency' columns. Choose the number of clusters to segment your customers.",
)

# Launch the Gradio app
interface.launch()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import gradio as gr

# Movie Recommendation Function
def recommend_movies(file, num_clusters, user_id):
    try:
        # Load the data from the uploaded file with the correct encoding
        data = pd.read_csv(file.name, encoding='ISO-8859-1')  # Adjust encoding if needed

        # Check for necessary columns
        required_columns = ['UserID', 'MovieTitle', 'Rating']
        if not all(col in data.columns for col in required_columns):
            return "Error: The file must contain 'UserID', 'MovieTitle', and 'Rating' columns."

        # Pivot the data to create a user-movie matrix
        user_movie_matrix = data.pivot_table(index='UserID', columns='MovieTitle', values='Rating', fill_value=0)

        # Normalize the data
        scaler = StandardScaler()
        scaled_matrix = scaler.fit_transform(user_movie_matrix)

        # Apply K-means clustering
        kmeans = KMeans(n_clusters=num_clusters, random_state=42)
        user_clusters = kmeans.fit_predict(scaled_matrix)

        # Assign clusters to the original dataset
        cluster_labels = pd.DataFrame({'UserID': user_movie_matrix.index, 'Cluster': user_clusters})
        data = pd.merge(data, cluster_labels, on='UserID')

        # Find the cluster of the given user
        if user_id not in cluster_labels['UserID'].values:
            return f"Error: User ID {user_id} is not found in the dataset."

        user_cluster = cluster_labels[cluster_labels['UserID'] == user_id]['Cluster'].iloc[0]

        # Find other users in the same cluster
        cluster_members = cluster_labels[cluster_labels['Cluster'] == user_cluster]['UserID']

        # Recommend the most popular movies among users in the same cluster
        cluster_data = data[data['UserID'].isin(cluster_members)]
        recommended_movies = (
            cluster_data.groupby('MovieTitle')['Rating'].mean().sort_values(ascending=False).head(5)
        ).index.tolist()

        return f"Recommended Movies for User {user_id}: {', '.join(recommended_movies)}"
    except Exception as e:
        return f"An error occurred: {str(e)}"

# Gradio Interface
file_input = gr.File(label="Upload CSV File (User-Movie Ratings)")
num_clusters_input = gr.Slider(2, 10, step=1, value=3, label="Number of Clusters")
user_id_input = gr.Number(label="Enter User ID for Recommendation")
recommendations_output = gr.Textbox(label="Recommended Movies")

interface = gr.Interface(
    fn=recommend_movies,
    inputs=[file_input, num_clusters_input, user_id_input],
    outputs=recommendations_output,
    title="Movie Recommendation AI",
    description="Upload a CSV file with 'UserID', 'MovieTitle', and 'Rating' columns. Enter a User ID to get personalized movie recommendations.",
)

# Launch the Gradio app
interface.launch()
```

```python
import gradio as gr
from PIL import Image

def compress_image(image, num_colors):
    try:
        # Open the image using PIL
        img = Image.open(image)
        # Convert the image to "P" mode for color quantization
        img = img.convert("P", palette=Image.ADAPTIVE, colors=num_colors)
        return img
    except Exception as e:
        return f"Error processing the image: {e}"

# Create the Gradio interface with a slider to control the number of colors
interface = gr.Interface(
    fn=compress_image,                     # Function to compress the image
    inputs=[
        gr.Image(type="filepath"),         # Input: Filepath of the image
        gr.Slider(1, 300, step=1, value=128, label="Number of Colors")  # Slider for color adjustment
    ],
    outputs=gr.Image(type="pil"),          # Output: Image with reduced colors
    title="Image Color Compression Tool",  # Add title to the interface
    description="Upload an image and reduce its colors using the slider."  # Description
)

# Launch the interface
interface.launch()
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = datasets.load_iris()
data = iris.data
target = iris.target  # Known species labels (for evaluation purposes)

# Apply KMeans Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(data)

# Predicted cluster labels
clusters = kmeans.labels_

# Create a DataFrame for visualization
df = pd.DataFrame(data, columns=iris.feature_names)
df['Cluster'] = clusters
df['Species'] = target  # Known labels for comparison

# Visualize clusters with Seaborn
sns.scatterplot(x=df['sepal length (cm)'], y=df['sepal width (cm)'], hue=df['Cluster'], palette='viridis')
plt.title("KMeans Clustering of Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# Evaluate the clustering performance
accuracy = sum(df['Cluster'] == df['Species']) / len(df)  # Quick comparison (not a formal metric)
print(f"Clustering accuracy (approximate): {accuracy:.2f}")
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import gradio as gr

# Example data: social media users
data = {
    'Age': [25, 30, 35, 40, 22, 28],
    'EngagementRate': [0.8, 0.6, 0.75, 0.5, 0.9, 0.65],
    'InterestsScore': [4, 3, 5, 2, 4.5, 3.8]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to perform clustering and visualization
def cluster_users(n_clusters):
    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(df)

    # Add cluster labels to the DataFrame
    df['Cluster'] = kmeans.labels_

    # Visualize the clusters
    fig, ax = plt.subplots(figsize=(8, 6))
    scatter = ax.scatter(df['Age'], df['EngagementRate'], c=df['Cluster'], cmap='viridis', s=100)
    ax.set_xlabel('Age')
    ax.set_ylabel('Engagement Rate')
    ax.set_title('User Clusters')
    plt.colorbar(scatter, label='Cluster')

    # Return the DataFrame and the figure
    return df, fig

# Create the Gradio interface
interface = gr.Interface(
    fn=cluster_users,
    inputs=gr.Number(label="Number of Clusters"),
    outputs=["dataframe", "plot"],
    title="Social Media User Clustering",
    description="Enter the number of clusters to group users based on Age and Engagement Rate.",
)

# Launch the Gradio app
interface.launch()
```
